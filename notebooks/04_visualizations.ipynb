```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("../data/cleaned/longevity_index.csv")

# Sort and select top 10 countries
top10 = df.sort_values(by='longevity_score', ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='longevity_score', y='country', data=top10, palette='viridis')
plt.title("🏆 Top 10 Countries by Longevity Score", fontsize=14)
plt.xlabel("Longevity Score (0–1)")
plt.ylabel("Country")
plt.tight_layout()

# Save image
plt.savefig("../assets/top10_longevity.png")

# Show plot
plt.show()
```
